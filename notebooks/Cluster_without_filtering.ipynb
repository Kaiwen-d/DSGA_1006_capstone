Silhouette score selection

```python
import sklearn.metrics as metrics
import numpy as np
score = []
cluster = np.arange(2,6)
print(cluster)
for i in range(2,6):
    labels=KMeans(n_clusters=i).fit(embedding).labels_ #,init="k-means++",random_state=200
    score.append(metrics.silhouette_score(embedding,labels,metric="euclidean"))
num_clusters = cluster[np.argmax(np.array(score))]
clustering_model = KMeans(n_clusters=num_clusters)
clustering_model.fit(embedding)
cluster_assignment = clustering_model.labels_
```

```python
clustered_sentences = [[] for i in range(num_clusters)]
for sentence_id, cluster_id in enumerate(cluster_assignment):
    clustered_sentences[cluster_id].append(sentences[sentence_id])

for i, cluster in enumerate(clustered_sentences):
    print("Cluster ", i+1)
    print(cluster)
    print("")
```

```python
embedding_all = []
for r in range(10):
  sentences = sent_tokenize(train_doc[r]) #split sentences
  embedding = []
  for sentence in sentences:
    embedding.append(model.encode(sentence, device='cuda')) #get embedding 
  embedding_all.append(embedding)
```

```python
num_cluster_all = []
cluster_assignment_all = []
silhouette_score_all = []
for r in range(10):
  sentences = sent_tokenize(train_doc[r]) #split sentences
  embedding = embedding_all[r] #get embedding 
  score = []
  cluster = np.arange(2,6)
  for i in range(2,6): # get best num_clusters from silhouette score
    labels=KMeans(n_clusters=i).fit(embedding).labels_ #,init="k-means++",random_state=200
    score.append(metrics.silhouette_score(embedding,labels,metric="euclidean"))
  silhouette_score_all.append(np.max(score))
  num_clusters = cluster[np.argmax(np.array(score))]
  clustering_model = KMeans(n_clusters=num_clusters)
  clustering_model.fit(embedding)
  cluster_assignment = clustering_model.labels_
  num_cluster_all.append(num_clusters)
  cluster_assignment_all.append(cluster_assignment)
```
